```python
import pandas as pd
import numpy as np

X = train.filter(regex='X')
y = train.filter(regex='Y')

# 숫자 변환 및 NaN/Inf 제거
X_clean = X.apply(pd.to_numeric, errors='coerce').replace([np.inf, -np.inf], np.nan)
y_clean = y['Y_01'].replace([np.inf, -np.inf], np.nan)

# 공통 유효 인덱스만 추출
valid_idx = X_clean.dropna().index.intersection(y_clean.dropna().index)
X_final = X_clean.loc[valid_idx]
y_final = y_clean.loc[valid_idx]

# 상관계수 계산
corr_series = X_final.corrwith(y_final).abs()
top_feature = corr_series.idxmax()
top_value = corr_series.max()

print(f"Y_01과 가장 높은 상관계수를 가진 변수: {top_feature}, 상관계수: {top_value:.4f}")
```

```python
from sklearn.feature_selection import VarianceThreshold

# 결측치가 있는 행 제거
X_clean = X.apply(pd.to_numeric, errors='coerce').replace([np.inf, -np.inf], np.nan).dropna()

# VarianceThreshold로 분산 측정
selector = VarianceThreshold(threshold=0.0001)
selector.fit(X_clean)

# 제거 대상 변수 목록
low_variance_features = X_clean.columns[~selector.get_support()].tolist()
print("❌ 분산이 매우 작은 변수들:", low_variance_features)
```

```python
# 각 변수의 고유값 개수 확인
unique_counts = X.nunique()

# 고유값이 2개 이하인 변수 추출
low_unique_features = unique_counts[unique_counts <= 2].index.tolist()

print("❌ 고유값이 2개 이하인 변수들:", low_unique_features)
```
